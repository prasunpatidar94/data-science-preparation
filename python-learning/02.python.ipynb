```python
# Comperision opration
ename="prasun"
eheight="5.50"
esal=200000000
ename1="Raja"
eheight1="5.50"
esal1=200000000

print("== with Str",ename==ename1)
print("== with float",eheight==eheight1)
print("== with number",esal==esal1)


print("!= with Str",ename!=ename1)
print("!= with float",eheight!=eheight1)
print("!= with number",esal!=esal1)

print("< with Str",ename<ename1)
print("<= with float",eheight<=eheight1)
print("< with number",esal<esal1)

print("> with Str",ename>ename1)
print("> with float",eheight>eheight1)
print(">= with number",esal>=esal1)

```
